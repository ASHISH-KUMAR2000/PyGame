```python
import pygame
import time
import random
pygame.init()

display_width = 800
display_height = 600

gameDisplay = pygame.display.set_mode((display_width, display_height))

pygame.display.set_caption('Saap Paal rkhe hai')

clock = pygame.time.Clock()

font = pygame.font.SysFont(None, 25)

def snake(block_size, snakeList):
    for element in snakeList:
        gameDisplay.fill((0, 0, 0), rect=[element[0], element[1], block_size, block_size])

def message_to_screen(msg, color):
    screen_txt = font.render(msg, True, color)
    gameDisplay.blit(screen_txt, [display_width//2, display_height//2])
    

def gameLoop():
    gameExit = False
    gameOver = False
    lead_x = display_width//2
    lead_y = display_height//2

    lead_x_change = 0
    lead_y_change = 0
    block_size = 10
    
    snakeList = []
    snakeLength = 1
    
    apple_size = 20
    apple_x = random.randrange( 0, display_width-apple_size, 10)
    apple_y = random.randrange( 0, display_height-apple_size, 10)
    
    
    
    while not gameExit:
        
        
        while gameOver == True:
            gameDisplay.fill((255, 255, 255))
            message_to_screen("You lose: Press q to quit or c to continue", (255, 0, 0))
            pygame.display.update()
        
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    gameExit = True
                    gameOver = False
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        gameOver = False
                        gameExit = True
                    elif event.key == pygame.K_c:
                        gameLoop()

        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                gameExit = True

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    lead_x_change = -block_size
                    lead_y_change = 0
                elif event.key == pygame.K_RIGHT:
                    lead_x_change = block_size
                    lead_y_change = 0
                elif event.key == pygame.K_UP:
                    lead_x_change = 0
                    lead_y_change = -block_size
                elif event.key == pygame.K_DOWN:
                    lead_x_change = 0
                    lead_y_change = block_size
        lead_x+=lead_x_change
        lead_y+=lead_y_change
        
        snakeList.append([lead_x, lead_y])
        
        #print(snakeList)
        if len(snakeList) > snakeLength:
            del snakeList[0]
            
            
        for element in snakeList[:-1]:
            if element == [lead_x, lead_y]:
                gameOver = True
                gameDisplay.fill((255, 255, 255))
                message_to_screen("Cross Over", (255, 0, 0))
                pygame.display.update()
                time.sleep(2)
                
        snake(block_size, snakeList)
        pygame.display.update()
        
        if lead_x >= apple_x and lead_x <= apple_x + apple_size:
            if lead_y >= apple_y and lead_y <= apple_y + apple_size:
                apple_x = random.randrange( 0, display_width-apple_size, 10)
                apple_y = random.randrange( 0, display_height-apple_size, 10)
                snakeLength+=1

        if lead_x >= display_width or lead_x < 0 :
            gameOver = True
        if lead_y < 0 or lead_y >= display_height :
            gameOver = True


        gameDisplay.fill((255, 255, 255))
        gameDisplay.fill((255, 0, 0), rect =[apple_x, apple_y, apple_size, apple_size])
        #gameDisplay.fill((0, 0, 0), rect=[lead_x, lead_y, 10, block_size])
        #pygame.display.update()
        clock.tick(30)

            
    pygame.quit()
    quit()
    
gameLoop()
```
